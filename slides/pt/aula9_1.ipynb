```ocaml

let l0 = []

let l1 = [1;2;3]

let l2 = 0::l1

let rec length l = 
  match l with
  | [] -> 0
  | _::xs -> 1 + length xs

let rec sum l = 
  match l with
  | [] -> 0
  | x::xs -> x + sum xs

let _ = List.fold_right (fun x ih -> x + ih) [1;2;3;4] 0

(* let rec induction l = 
  match l with
  | [] -> base_case
  | x::xs -> let ih = induction xs in inductive_case x ih *)


let rec fold_right f l bc = 
  match l with
  | [] -> bc
  | x::xs -> let ih = fold_right f xs bc in f x ih

let _ = assert (length [1;2;3] = 3)
let _ = assert ((fold_right (fun _ ih -> 1+ih ) [1;2;3] 0) = 3)
let _ = assert ((fold_right (fun x ih -> x+ih ) [1;2;3] 0) = 6)
```

```ocaml
let rec fold_left f acc l = 
  match l with
  | [] -> acc
  | x::xs -> fold_left f (f acc x) xs 

let _ = assert (fold_left (fun w x -> w+x) 0 [1;2;3] = 6)

let _ = assert (List.fold_left (fun w x -> w+x) 0 [1;2;3] = 6)

let _ = assert (List.fold_left (+) 0 [1;2;3] = 6)
let _ = assert (List.fold_left ( * ) 1 [1;2;3;2] = 12)
```
