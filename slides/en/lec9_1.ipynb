```ocaml
let rec count_by_group l = 
  match l with 
  | [] -> []
  | x::xs -> 
      let ps = count_by_group xs in 
      match ps with
      | [] -> [(x,1)]
      | (y,n)::ps' -> if x = y then (y,n+1)::ps' else (x,1)::ps

let _ = (count_by_group ["a"] = [("a",1)])
let _ = (count_by_group ["a";"a";"a";"b";"b";"c";"a"] = [("a",3);("b",2);("c",1);("a",1)])
let _ = (count_by_group ("a"::["a";"a";"a";"b";"b";"c";"a"]) = [("a",4);("b",2);("c",1);("a",1)])
let _ = (count_by_group ("b"::["a";"a";"a";"b";"b";"c";"a"]) = [("b",1);("a",3);("b",2);("c",1);("a",1)])
```

```ocaml
let rec length l = 
  match l with 
  | [] -> 0
  | h::t -> 1 + length t

let rec sumL l = 
  match l with 
  | [] -> 0
  | h::t -> h + sumL t

let rec fold_right f l base_case = 
  match l with
  | [] -> base_case
  | x::xs -> let ih = fold_right f xs base_case in f x ih

let _ = assert (length [1;2;3] = 3)
let _ = assert (fold_right (fun _ ih -> 1+ih) [1;2;3] 0 = 3)
let _ = assert (fold_right (fun x ih -> x+ih) [1;2;3] 0 = 6)
let _ = assert (List.fold_right (fun _ ih -> 1+ih) [1;2;3] 0 = 3)
let _ = assert (List.fold_right (fun x ih -> x+ih) [1;2;3] 0 = 6)
```

```ocaml
let rec length_tr l acc = 
  match l with 
  | [] -> acc
  | h::t -> length_tr t (1 + acc)

let rec sumL_tr l acc = 
  match l with 
  | [] -> acc
  | h::t -> sumL_tr t (h + acc)

let rec fold_left f acc l = 
  match l with
  | [] -> acc
  | h::t -> let next_acc = f acc h in fold_left f next_acc t


let _ = assert (fold_left (fun ih _  -> 1+ih) 0 [1;2;3]  = 3)
let _ = assert (fold_left (fun ih x  -> x+ih) 0 [1;2;3]  = 6)
let _ = assert (List.fold_left (fun ih _  -> 1+ih) 0 [1;2;3]  = 3)
let _ = assert (List.fold_left (fun ih x  -> x+ih) 0 [1;2;3]  = 6)
```

```ocaml
let l0 = [1;2;3;4;5;6]
let l1 = [2;4;6]

let rec filter f l = 
  match l with
  | [] -> []
  | x::xs -> 
    let fs = filter f xs in 
      if f x then x::fs else fs

let filter' f l = List.fold_right (fun x fs -> if f x then x::fs else fs) l []

let _ = assert (filter (fun x -> x mod 2 = 0) l0 = l1)
let _ = assert (filter' (fun x -> x mod 2 = 0) l0 = l1)

```
