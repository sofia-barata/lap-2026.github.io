````ocaml
(* 
    [add f1 f2] devolve a adição das frações [f1] e [f2]
    requires: [f1.denom > 0 && f2.denom > 0] 
*)
let add_frac f1 f2 =
  simp {num = (f1.num*f2.den + f2.num*f1.den); den = f1.den*f2.den}

let _ = add_frac (frac 1 2) (frac 2 4)
````

````ocaml
(* 
    [neg_frac f] devolve o oposto da fração [f]
    requires: [f.denom > 0] 
*)
let neg_frac {num;den} = frac (-num) den

let _ = neg_frac (frac 1 2)

````

````ocaml
(*
    [sub_frac f1 f2] devolve a subtração das frações [f1] e [f2]
    requires: [f1.denom > 0 && f2.denom > 0] 
*)
let sub_frac f1 f2 =
  add_frac f1 (neg_frac f2)

let _ = sub_frac (frac 1 2) (frac 1 2)
````
